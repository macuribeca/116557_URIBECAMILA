```python
'''Leer una lista de números enteros y determinar cual es el segundo elemento 
mayor de dicha lista. No utilizar ningún tipo de algoritmo de ordenamiento para 
realizar este ejercicio. Por ejemplo, en la lista [11, 22, 3, 4, 22], el 
segundo número mayor sería el 11; de otro lado, en la lista [1, 1, 1, 1, 1] no 
existe un segundo elemento mayor.'''

def leer_lista_numeros():

    while True:
        try:
            lista= list(eval(input("Ingerese la lista de números separando cada \
                                  termino con comas ")))
            for x in range(len(lista)-1):
                if type(lista[x]) != int:
                    raise ValueError
            break
        except:
            print("Oops... Los números ingresados a la lista no están ordenados\
                   ascendentemente")
    return(lista)
   

lista = leer_lista_numeros()
numero_mayor_1 = -float("inf")
numero_mayor_2 = -float("inf")
for x in lista:
    if x > numero_mayor_1:
        numero_mayor_1 = x
lista.remove(numero_mayor_1)
for x in lista:
    if x > numero_mayor_2:
        numero_mayor_2 = x
print (f"El segundo número mayor de la lista es {numero_mayor_2}")
```

```python
'''Leer una matriz e imprimir el vector que aparece cuando la matriz se 
desenreda en forma de espiral que gira en el sentido horario a partir de la 
esquina inferior derecha.'''

M = [] # vector vacío

filas = int(input('Ingrese el número de filas de la matriz: '))
columnas = int(input('Ingrese el número de columnas de la matriz: '))

for i in range(filas): 
    M.append([]) #Se añade []
    for j in range(columnas):
        M[i].append(int(input(f'Escriba el elemento #{j+1} de la fila {i+1}: \
         '))) #Se agrega el valor ingresado a la sublista correspondiente

num_elementos = filas*columnas #Cantidad de elementos de la matriz
vec = []
col_final = columnas-1
col_inicial = 0
fil_final = filas-1
fil_inicial = 0
cont=0

while cont < num_elementos: 
    for i in range(col_final,col_inicial,-1): 
        pos = M[fil_final][i] 
        vec.append(pos)
        cont +=1 
    
    for i in range(fil_final,fil_inicial,-1):
        pos = M[i][col_inicial]
        vec.append(pos)
        cont +=1 

    for i in range(col_inicial,col_final):#
        pos = M[fil_inicial][i]
        vec.append(pos)
        cont +=1 
    
    for i in range(fil_inicial,fil_final):
        pos = M[i][col_final]
        vec.append(pos)
        cont +=1
    
    col_final -= 1
    col_inicial += 1
    fil_final = filas - 1
    fil_inicial += 1

print(vec)
```

```python
'''Diseñe una función de tres parámetros que pida la altura y anchura de un 
rectángulo, y el carácter a utilizar para dibujarlo.'''

def fun_rectángulo(altura, anchura, carácter):
    for i in range(altura):
        for j in range(anchura):
            print(carácter, end=" ")
        print()

while True:
    try:
        altura   = int(input("Digite la altura del rectángulo: "))
        anchura  = int(input("Digite la anchura del rectángulo: "))
        break
    except ValueError:
        print("Debe ingresar valores enteros.")

carácter = input("Carácter a utilizar: ")

fun_rectángulo(altura, anchura, carácter)
```

```python
'''Diseñe una función que pida la anchura de un triángulo y lo dibuje con 
caracteres producto (*).'''

def fun_triángulo_1(anchura):        #Función triángulo creciente
    for i in range(1, anchura + 1):
        for j in range(i):
            print("* ", end="")
        print()

def fun_triángulo_2(anchura):        #Función triángulo decreciente
    for i in range(1, anchura):
        for j in range(anchura - i):
            print("* ", end="")
        print()

while True:
    try:
        anchura = int(input("Digite la anchura del triángulo: "))
        if anchura >= 0:
            break
        else:
            print("Debe ingresar valores positivos. Intentelo nuevamente.")
    except ValueError:
        print("Debe ingresar valores enteros. Intentelo nuevamente.")

fun_triángulo_1(anchura)
fun_triángulo_2(anchura)
```

```python
'''Diseñe una función que pida dos años y escriba cuántos años bisiestos hay 
entre esas dos fechas (incluidos los dos años).'''

def es_bisiesto(t):
    return t % 400 == 0 or (t % 100 != 0 and t % 4 == 0)

print("Contador de años bisiestos")
inicio = int(input("Escriba un año: "))
print("Escriba otro año posterior a", str(inicio) + ": ", end="")
final = int(input())
while final < inicio:
    print(final, "no es mayor que", str(inicio) + ". Inténtelo de nuevo: ",
          end="")
    final = int(input())

contador = 0
for i in range(inicio, final + 1):
    if es_bisiesto(i):
        contador += 1

print("De", inicio, "a", final, "hay", final - inicio + 1, "año(s),",
      contador, "de ellos bisiesto(s).")
```
